```python
import scipy.io as sio

from utils import *

# Load the data
depth_results = sio.loadmat('data/results_depth.mat')
depth_results = depth_results['results']
width_results = sio.loadmat('data/results_width.mat')
width_results = width_results['results']
```

## 1. Display the dependence of training time on network depth

```python
df = results_to_dataframe(depth_results, variable_name="Layers", matlab_results=True)
# df = df.where(df["Units"] != 10).dropna()

title = "Matlab Mean Training Time vs. Number of Layers for Different LSTM Cell Types"
xlabel = "Number of Layers"
plot_training_time(df, variable_name="Layers", use_bars=True, bar_width=0.9, rotation=0, title=title, xlabel=xlabel)
plot_training_time(df, variable_name="Layers", use_bars=False, rotation=0, title=title, xlabel=xlabel)
```

## 2. Display dependence of training time on networks width (units)

```python
df = results_to_dataframe(width_results, variable_name="Units", matlab_results=True)
df = df.where(df["Units"] != 10).dropna()

title = "Matlab Mean Training Time vs. Number of Hidden Units for Different Cell Types"
plot_training_time(df, variable_name="Units", use_bars=True, bar_width=25.0, rotation=0, title=title)
plot_training_time(df, variable_name="Units", use_bars=False, rotation=0, title=title)
```
