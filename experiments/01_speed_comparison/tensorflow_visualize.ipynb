```python
import numpy as np
import pickle
import pandas as pd
import matplotlib.pyplot as plt

from utils import *
```

## 1. Display the dependence of training time on network depth

```python
with open("data/tensorflow_depth_training_time.pkl", "rb") as f:
    experiment = pickle.load(f)
results = experiment["Results"]

df = results_to_dataframe(results, variable_name="Layers")
# df = df.where(df["Units"] != 10).dropna()

title = "Tensorflow Mean Training Time vs. Number of Layers for Different LSTM Cell Types"
xlabel = "Number of Layers"
plot_training_time(df, variable_name="Layers", use_bars=True, bar_width=0.9, rotation=0, title=title, xlabel=xlabel)
plot_training_time(df, variable_name="Layers", use_bars=False, rotation=0, title=title, xlabel=xlabel)
```

## 2. Display dependence of training time on networks width (units)

```python
with open("data/tensorflow_width_training_time.pkl", "rb") as f:
    experiment = pickle.load(f)	
results = experiment["Time Measurements"]

df = results_to_dataframe(results, variable_name="Units")
df = df.where(df["Units"] != 10).dropna()
plot_training_time(df, variable_name="Units", use_bars=True, bar_width=25.0, rotation=0)
plot_training_time(df, variable_name="Units", use_bars=False, rotation=0)
```
